```python
movies = movies.merge(credits,on="title")             # Merging both the datasets
```

Fetching important columns from the data

```python
movies.info()
```

```python
movies = movies[["movie_id","title","overview","genres","keywords","cast","crew"]]
```

```python
movies.head()
```

```python
movies.shape
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
import ast     # To convert it into proper list format.
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies.head()
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i["name"])
            counter+=1
        else:
            break
    return L
```

```python
movies["cast"] = movies["cast"].apply(convert3)
```

```python
movies.head()
```

```python
def director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
            break
    return L
```

```python
movies["crew"] = movies["crew"].apply(director)
```

```python
movies.head()
```

```python
movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
movies["tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
movies.head()
```

```python
new_df = movies[["movie_id","title","tags"]]
```

```python
new_df
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df["tags"] = new_df["tags"].apply(stem)
```

```python
new_df["tags"]
```

```python
new_df["tags"][0]
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:x.lower())
```

```python
new_df
```

# Now , we will convert text into vectors i.e. called Text Vectorization.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
len(cv.get_feature_names_out())
```

```python
new_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)     # Calculate the distance between the vectors
```

```python
similarity
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df["title"] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Batman Begins")                   # Related movies will be recommended
```
